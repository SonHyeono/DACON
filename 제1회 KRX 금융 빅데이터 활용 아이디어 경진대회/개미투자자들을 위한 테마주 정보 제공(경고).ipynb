```python
from selenium import webdriver
from selenium.webdriver.common.by import By
driver = webdriver.Chrome(executable_path='./chromedriver.exe')
```

```python
from urllib.parse import quote
import requests
# url 접속
query = '테마주'
url = "https://search.naver.com/search.naver?where=view&sm=tab_jum&query=" + quote(query)
driver.get(url)
# 1초 대기
time.sleep(1)
```

```python
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.93 Safari/537.36"}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "lxml")
li = soup.find_all("li", attrs={"class":"bx _svp_item"})
```

```python
theme_txt = []
for txt in li:
    try:
        txt_title = txt.find("a", attrs={"class":"api_txt_lines total_tit _cross_trigger"}).get_text()
        print("제목 :",txt_title)
        txt_short = txt.find("a", attrs={"class":"total_dsc _cross_trigger"}).get_text()
        print("요약 :", txt_short)
        txt_tag = txt.find("div", attrs={"class":"total_tag_area"}).get_text()
        print("태그 :", txt_tag)
        txt_child = txt.find("div", attrs={"class":"total_link_area"}).get_text()
        print("자식 :", txt_child)
        
        theme_txt.append(txt_title + " " + txt_short + " " + txt_tag + " " + txt_child)
    except:
        print("none")
    print("=========================================")
    
    
```
