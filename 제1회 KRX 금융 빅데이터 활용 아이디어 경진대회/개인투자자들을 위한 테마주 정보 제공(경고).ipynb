```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import pandas as pd
pd.set_option('display.max_columns', None) # 데이터프레임 컬럼 display

import warnings
warnings.filterwarnings('ignore') # 경고 무시
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
driver = webdriver.Chrome(executable_path='./chromedriver.exe')
```

```python
from urllib.parse import quote
import requests
# url 접속
query = '테마주'
url = "https://search.naver.com/search.naver?where=view&sm=tab_jum&query=" + quote(query)
driver.get(url)
# 1초 대기
time.sleep(1)
```

```python
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.93 Safari/537.36"}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "lxml")
li = soup.find_all("li", attrs={"class":"bx _svp_item"})
```

```python
theme_txt = []
for txt in li:
    try:
        txt_val = txt.find("a", attrs={"class":"api_txt_lines total_tit _cross_trigger"}).get_text()
        txt_val += txt.find("a", attrs={"class":"total_dsc _cross_trigger"}).get_text()
        txt_val += txt.find("div", attrs={"class":"total_tag_area"}).get_text()
        txt_val += txt.find("div", attrs={"class":"total_link_area"}).get_text()
    except:
        print("none")
    
    print("블로그 :", txt_val)
    theme_txt.append(txt_val)
    
    print("=========================================")
    
    
```

```python
import re

theme_list = []

for txt in theme_txt:
    txt = re.sub('[^가-힣a-zA-Z0-9 ]',' ',txt)
    txt_list = txt.split(" ")
    
    for i, t in enumerate(txt_list):
        if t == '테마주':
            if i != 0:
                theme_list.append(txt_list[i - 1] + " 테마주")
        elif t == '관련주':
            if i != 0:
                theme_list.append(txt_list[i - 1] + " 관련주")
        elif '테마주' in t:
            if t.find('테마주') == 0:
                theme_list.append(txt_list[i - 1] + " 테마주")
            else:
                theme_list.append(txt_list[i])
        elif '관련주' in t:
            if t.find('관련주') == 0:
                theme_list.append(txt_list[i - 1] + " 관련주")
            else:
                theme_list.append(txt_list[i])
```

```python
theme_list = list(set(theme_list))
theme_list
```
