```python
import pandas as pd
```

```python
df_train = pd.read_csv('open/train.csv')
df_val = pd.read_csv('open/val.csv')
```

```python
df_val
```

## df_val : 정상 0, 사기 1

```python
val_one = df_val[df_val['Class'] == 1]
```

```python
feature_names = ['V' + str(i) for i in range(1, 31)]
```

```python
feature = pd.DataFrame(df_val, columns=feature_names)
target = pd.DataFrame(df_val, columns=['Class'])
```

```python
total = pd.concat([feature,target], axis=1)
```

```python
total_corr = total.corr()
```

```python
total_corr = total_corr['Class']
```

```python
total_corr = pd.DataFrame(total_corr.values, index= total_corr.index, columns = ['Corr'])
```

```python
total_corr.sort_values('Corr')
```

```python
# 0.2 면 0.8186727166
df = total_corr[abs(total_corr['Corr']) > 0.1]
df
```

```python
df.index
```

```python
df_col = df_val[df.index]
```

```python
df_col
```

```python
from sklearn.neural_network import MLPClassifier
model = MLPClassifier(max_iter=500)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(df_col.drop('Class', axis=1), df_col['Class'], test_size = 0.2)
```

```python
model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score

model.score(X_train, Y_train), model.score(X_test, Y_test)
```
